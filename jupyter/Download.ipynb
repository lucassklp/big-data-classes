```python
import urllib.request
import datetime
import shutil
import os
import gzip

start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2021, 1, 31)
delta = datetime.timedelta(days=1)

def gunzip(source_filepath, dest_filepath, block_size=65536):
    with gzip.open(source_filepath, 'rb') as s_file, \
            open(dest_filepath, 'wb') as d_file:
        while True:
            block = s_file.read(block_size)
            if not block:
                break
            else:
                d_file.write(block)



while start_date <= end_date:
    date = str(start_date)
    filename = date + ".ndjson"
    filenameGz = filename + ".gz"
    url = "https://openaq-fetches.s3.amazonaws.com/dbbackup/" + date + "/" + filenameGz
    print(url)
    urllib.request.urlretrieve(url, filenameGz)
    gunzip(filenameGz, "dataset/"+filename)
    os.remove(filenameGz)
    start_date += delta

```
